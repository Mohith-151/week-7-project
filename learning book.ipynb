## Hypothesis testing
- **NULL hypothesis**
  - a default assumption
- **Althernative hypothesis**
  - a statement or a goal to prove something
  - if succeed we can reject the NULL hypothesis, which could be good or bad based on default assumption.
  - *evidence* is a key term in alternative hypothesis.

## correlation & covariance.
*both measures how two variable are related to one other.*

- **Correlation**
    - tells the directiona & strength of linear relationship
    - ranges: -1 to +1. 
        - where +1 means perfect positive correlation , 0 means no correlation and -1 perfect negative correlation
    
- **Covariance**
    - tells the direction of linear relationship
    - positive value indicates the variables are moving together and negative means inverse

## Statistical significance and p-values

- **p-value** : a probabily value got by observing the data, usually it assumes the NULL hypothesis is true.
- **Statistical significance** : if p-value is lower than the predefined threshold(usually 0.05), we can reject NULL hypothesis. 

## Confidence Intervals & Margin of error.

By researching about in depth, i understand that this topic  is very much imp. in the data science.As these are the final result or statement the CEO or the upper officers wants from a Data scientist.

In general this means:
- **margin of error:**
    - a percentage calculated on the dataset.
    - which tells about 'How much u expect that our data(sample data) is differ from population(the large data or the real world.)?'

- **confidence intervals**
    - it is range calculated using the margin of error percentage.
    - which tells 'how much confidence you have that the predicted result is true(give in range of percentage, for example: 47% - 53%)
    -
    -
    The final statement we(the data scientist would give) after performing all required process will look like.
    - **"Candidate A is leading with 50% of the vote, with a margin of error of ±3%."**
    - **In the context of whole country's voter we are confident that about 43% - 53% of voters will vote for Candidate A**


to calculate the confidence interval.....

```python
import scipy.stats as stats
import numpy  as np

confidence_level = 0.95
degrees_freedom = len(df['Daily_Revenue']) - 1
sample_mean = np.mean(df['Daily_Revenue'])
sample_standard_error = stats.sem(df['Daily_Revenue'])  #<--- the margin of error

confidence_interval = stats.t.interval(confidence_level, degrees_freedom, sample_mean, sample_standard_error)

print(f"We are 95% confident that the true average revenue lies between {confidence_interval[0]:.2f} and {confidence_interval[1]:.2f}")
```
